```python
# Combine 'Date' and 'Time' columns into a single DateTime column
df['DateTime'] = pd.to_datetime(df['Date'] + ' ' + df['Time'])

# Set the DateTime column as the DataFrame index
df.set_index('DateTime', inplace=True)
```

```python
# Resample the data to daily frequency and calculate total sales per day
daily_sales = df['Total'].resample('D').sum()
```

```python
plt.figure(figsize=(12, 6))
plt.plot(daily_sales.index, daily_sales.values)
plt.xlabel('Date')
plt.ylabel('Total Sales')
plt.title('Total Daily Sales Over Time')
plt.grid(True)
plt.show()
```

```python
from statsmodels.tsa.seasonal import STL
decomposition = STL(daily_sales, seasonal=13)
result = decomposition.fit()
result.plot()
plt.show()
```

```python
from statsmodels.tsa.stattools import adfuller
adf_result = adfuller(daily_sales)
print(f'ADF Statistic: {adf_result[0]}')
print(f'p-value: {adf_result[1]}')
```

```python
# First-order differencing
stationary_sales = daily_sales.diff().dropna()
```

```python
plt.figure(figsize=(12, 6))
plt.plot(stationary_sales.index, stationary_sales.values)
plt.xlabel('Date')
plt.ylabel('Total Sales')
plt.title('Stationary Total Daily Sales Over Time')
plt.grid(True)
plt.show()
```

```python
from statsmodels.graphics.tsaplots import plot_acf, plot_pacf

# Create subplots for ACF and PACF plots
fig, (ax1, ax2) = plt.subplots(2, 1, figsize=(12, 8))

# Plot ACF
plot_acf(stationary_sales, lags=30, ax=ax1)
ax1.set_title('Autocorrelation Function (ACF) Plot')
ax1.set_xlabel('Lag')
ax1.set_ylabel('ACF Value')

# Plot PACF
plot_pacf(stationary_sales, lags=30, ax=ax2)
ax2.set_title('Partial Autocorrelation Function (PACF) Plot')
ax2.set_xlabel('Lag')
ax2.set_ylabel('PACF Value')

plt.tight_layout()
plt.show()
```
