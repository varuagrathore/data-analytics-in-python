```python
import pandas as pd
```

```python
df=pd.read_csv('sales_data_Ar.csv')
df
```

```python
grouped = df.groupby('catégorie')
```

```python
grouped.mean(numeric_only=True)
#if you not use this condition in the bracket then you get error or a warning that warning not create any problem because 
#here it can only calculate mean of numeric column here non numeric coloumn is not involve
```

```python
grouped.count()
```

```python
result = grouped['Quantity Ordered'].agg(['sum', 'mean', 'median', 'max', 'min'])
result

```

```python
vetements are in high demand because its value of sum is greater then others 
and sports are in lower demands .

On average, customers tend to order around 1.07 to 1.15 items in all categories. However, "Alimentation" and "Vêtements" 
have slightly higher average order quantities compared to "Sports" and "Électronique."

In all categories, the median order quantity is 1.0,
indicating that half of the orders in each category consist of a single item.

The 'max' values show the maximum order quantity in each category. "Vêtements" has
the highest maximum order quantity (4), indicating that some customers placed larger
orders in this category. "Sports" and "Électronique" have lower maximum order quantities.
```
