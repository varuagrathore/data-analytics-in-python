```python
import pandas as pd
import numpy as np
data = {
    'Name': [f'Name{i}' for i in range(1, 51)],
    'Age': np.random.randint(18, 65, size=50),
    'City': np.random.choice(['New York', 'Los Angeles', 'Chicago', 'Houston'], size=50),
    'Salary': np.random.randint(30000, 100000, size=50)
}

df = pd.DataFrame(data)
```

```python
df
```

Select all rows where 'Age' is greater than 30.

```python
AR=(df['Age']>30)
```

```python
df.loc[AR]
```

Select all rows where 'City' is 'New York'.

```python
df.loc[df['City']== 'New York']
```

Select all rows where 'Salary' is less than $50,000.

```python
df.loc[df['Salary']<50000]
```

Select all rows where 'Age' is between 25 and 35.

```python
res=(df['Age']>=25)&(df['Age']<=35)
df.loc[res]
```

Select the top 10 highest earners.

```python
sorted_df=df.sort_values('Salary',ascending=False)
```

```python
sorted_df
```

```python
sorted_df.head(10)
```

Count the number of people in each city.

```python
df['City'].value_counts()
```

Calculate the average age of people in 'Los Angeles'.

```python
df.loc[df['City']=='Los Angeles','Age'].mean()
```

Find the person with the highest salary.

```python
df['Salary'].max()
```

Select all rows where 'Name' starts with the letter 'N'.

```python
df.loc[df['Name'].str[0]=='N']
```

Calculate the total salary expenditure.

```python
df['Salary'].sum()
```

Select all rows where 'City' is not 'New York'.

```python
filt=(df['City']=='New York')
df.loc[~filt]
```

Calculate the median age of people in 'Chicago'.

```python
df.loc[df['City']=='Chicago','Age'].median()
```
